###### About Dataset
This is a countrywide car accident dataset that covers 49 states of the USA. The accident data were collected from February 2016 to March 2023, using multiple APIs that provide streaming traffic incident (or event) data. These APIs broadcast traffic data captured by various entities, including the US and state departments of transportation, law enforcement agencies, traffic cameras, and traffic sensors within the road networks. The dataset currently contains approximately 7.7 million accident records. For more information about this dataset, please visit here.

###### Dataset Information

* ID: A unique identifier for each accident.
* Severity: Indicates the impact or seriousness of the accident (often on a scale from 1 to 4, with 1 being minor and 4 being severe).
* Start_Time: The date and time when the accident occurred or was reported.
* End_Time: The date and time when the accident was resolved or cleared.
* Start_Lat: The latitude where the accident started.
* Start_Lng: The longitude where the accident started.
* End_Lat: The latitude where the accident was cleared (if applicable).
* End_Lng: The longitude where the accident was cleared (if applicable).
* Distance(mi): The distance (in miles) impacted by the accident.
* Description: A brief description of the accident or incident.
* Number of Lanes Affected: The number of lanes impacted by the accident.
* Street: The street where the accident occurred.
* City: The city where the accident happened.
* County: The county where the accident happened.
* State: The state where the accident took place.
* Zipcode: The postal code of the area where the accident occurred.
* Timezone: The time zone where the accident took place.
* Temperature(F): The temperature at the time of the accident.
* Wind_Chill(F): The wind chill factor during the accident.
* Humidity(%): The humidity level at the time of the accident.
* Pressure(in): Atmospheric pressure at the accident location.
* Visibility(mi): Visibility distance in miles at the time of the accident.
* Wind_Direction: The direction of the wind.
* Wind_Speed(mph): The speed of the wind during the accident.
* Precipitation(in): Amount of precipitation (rain/snow) recorded during the accident.
* Weather_Condition: A description of the weather (e.g., clear, rainy, foggy) at the time of the accident.
* Amenity: Indicates whether the accident occurred near a convenience amenity (e.g., gas stations, restaurants).
* Bump: Indicates whether there was a bump on the road (e.g., speed bumps).
* Crossing: Indicates if the accident occurred at or near a pedestrian crossing.
* Give_Way: Indicates if there was a "Give Way" sign near the accident.
* Junction: Indicates if the accident occurred near a road junction.
* No_Exit: Indicates if the accident happened in a no-exit zone.
* Railway: Indicates proximity to a railway crossing.
* Roundabout: Indicates if the accident occurred near or in a roundabout.
* Station: Indicates proximity to a station (bus or train station).
* Stop: Indicates if there was a stop sign near the accident.
* Traffic_Calming: Refers to measures like speed bumps or narrowed roads near the accident.
* Traffic_Signal: Indicates whether a traffic signal was present at the accident site.
* Turning_Loop: Indicates if the accident occurred near a turning loop.
* Sunrise_Sunset: Describes whether the accident occurred during daylight or nighttime.
* Civil_Twilight: Describes the lighting condition during the twilight period.
* Nautical_Twilight: Time during which the horizon is visible at sea.
* Astronomical_Twilight: The time when the sky is dark enough for all astronomical observations to begin.

```python
data_df.head()
```

###### Converting categorical columns to numerical columns

```python
data_df['Month'] = data_df['Month'].map({'January':0,'February':1,'March':2,
                                         'April':3,'May':4,'June':5,'July':6,'August':7,
                                         'September':8,'October':9,'November':10,'December':11})
data_df['Month_End'] = data_df['Month_End'].map({'January':0,'February':1,'March':2,
                                         'April':3,'May':4,'June':5,'July':6,'August':7,
                                         'September':8,'October':9,'November':10,'December':11})

data_df['Day'] = data_df['Day'].map({'Monday':0,'Tuesday':1,'Wednesday':2,'Thursday':3,'Friday':4,
                                     'Saturday':5,'Sunday':6})
data_df['Day_End'] = data_df['Day_End'].map({'Monday':0,'Tuesday':1,'Wednesday':2,'Thursday':3,'Friday':4,
                                             'Saturday':5,'Sunday':6})
```
